```python
import networkx as net
import matplotlib.pyplot as plt
import networkx.generators.small
```

```python
# Create the graph
g = networkx.generators.small.krackhardt_kite_graph()
```

```python
# rename nodes to names
names = ["Beverly", "Andre","Ed", "Diane","Carol", "Garth", "Fernando", "Heather","Ike","Jane"]
```

```python
# Change nodes from numbers to names
name_nodes = dict(zip(range(10),names))
```

```python
# rename the nodes
g = net.relabel_nodes(g, name_nodes)
```

```python
# change the layout
pos = net.spring_layout(g, seed=42)
# change axis from left to right
pos = {node: (-x, y) for node, (x, y) in pos.items()}
# Draw the graph
net.draw(g, with_labels= True)
```
