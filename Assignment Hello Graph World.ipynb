```python
import networkx as net
import matplotlib.pyplot as plt
import networkx.generators.small
```

```python
# Create the graph
g = networkx.generators.small.krackhardt_kite_graph()
```

```python
# rename nodes to names
names = ["Andre", "Beverly","Carol", "Diane","Ed", "Fernando", "Garth", "Heather","Ike","Jane"]
```

```python
# Change nodes from numbers to names
name_nodes = dict(zip(range(10),names))
```

```python
# rename the nodes
g = net.relabel_nodes(g, name_nodes)
```

```python
# Draw the graph
net.draw(g, with_labels= True)
```
